```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

from sklearn.cluster import KMeans
```

```python
raw_data = pd.read_csv('3.12. Example.csv')
```

```python
raw_data
```

```python
x = raw_data.copy()
x
```

```python
plt.scatter(raw_data['Loyalty'], raw_data['Satisfaction'])
plt.xlabel('Satisfaction')
plt.ylabel('Loyalty')
plt.show
```

```python
kmeans = KMeans(2)
kmeans.fit(x)
```

```python
cluster_pred = kmeans.fit_predict(x)
```

```python
x_pred = x.copy()
x_pred['Clusters'] = cluster_pred
```

```python
x_pred
```

```python
plt.scatter(x_pred['Satisfaction'],x_pred['Loyalty'], c = x_pred['Clusters'], cmap='rainbow')
plt.ylabel('Satisfaction')
plt.xlabel('Loyalty')
plt.show
```

## Standardize the Variables

```python
from sklearn import preprocessing
x_scaled = preprocessing.scale(x)
x_scaled
```

```python
wcss = []
for i in range(1,10):
    kmeans = KMeans(i)
    kmeans.fit(x_scaled)
    wcss_iter = kmeans.inertia_
    wcss.append(wcss_iter)
wcss
```

```python
plt.plot(range(1,10), wcss)
plt.xlabel('Number of Clusters')
plt.ylabel('wcss')
plt.show
```

## Explore Cluster Solutions to Select the Number of Clusters

```python
kmeans_new = KMeans(4)
kmeans_new.fit(x_scaled)
clusters = kmeans_new.fit_predict(x_scaled)
```

```python
clusters
```

```python
new_data = x.copy()
new_data['Clusters'] = clusters
new_data
```

```python
plt.scatter(new_data['Satisfaction'], new_data['Loyalty'], c = new_data['Clusters'], cmap = 'rainbow')
plt.xlabel('Satisfaction')
plt.ylabel('Loyalty')
plt.show
```
